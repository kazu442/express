```python
! curl -s -O http://matsuda.php.xdomain.jp/pe/rectangle.png
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

画像 = cv2.imread('rectangle.png') # 画像の読み込み
グレー = cv2.cvtColor(画像, cv2.COLOR_BGR2GRAY) # グレースケール変換
グレー = グレー.astype(np.float32) # 小数化
高さ, 幅 = グレー.shape
エッジ画像 = グレー.copy()

for y in range(高さ):
    for x in range(幅-1): # 1つ先のピクセルと引き算するので1引くことに注意
        エッジ画像[y, x]=グレー[y, x+1]-グレー[y, x]

エッジ画像[エッジ画像<0]=91
cv2_imshow(エッジ)
```

```python
! curl -s -O http://matsuda.php.xdomain.jp/pe/rectangle.png
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

画像 = cv2.imread('rectangle.png') # 画像の読み込み
グレー = cv2.cvtColor(画像, cv2.COLOR_BGR2GRAY) # グレースケール変換

xフィルタ = np.array([[0,  0, 0],
[0, -1, 1],
                   [0,  0, 0]])
xエッジ = cv2.filter2D(グレー, cv2.CV_32F, xフィルタ)  # フィルタ処理

yフィルタ = np.array([[0,  0, 0],
                   [0, -1, 0],
                   [0,  1, 0]])
yエッジ = cv2.filter2D(グレー, cv2.CV_32F, yフィルタ)  # フィルタ処理

エッジ画像 = np.sqrt(xエッジ ** 2 + yエッジ ** 2) # 距離計算
cv2_imshow(エッジ画像) # 結果の表示
```
